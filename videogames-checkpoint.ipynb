Finally, we will test whether average user ratings of the Xbox One and PC platforms are the same. We will also test whether average user ratings for the Action and Sports genres are different. We will do this using hypothesis testing. 


</div>

```python
#Null Hypothesis: Average user ratings of the Xbox One and PC platforms are the same. 
#Alternative Hypothesis: Average user ratings for Xbox One and PC are not the same. 

#XOne
xone_ratings = filtered_games.query("platform == 'XOne' and user_score.notna()")['user_score']

#PC
pc_ratings = filtered_games.query("platform == 'PC' and user_score.notna()")['user_score']

t_stat, p_value = ttest_ind(xone_ratings, pc_ratings)

alpha = 0.05

if p_value < alpha: 
    print("We reject the null hypothesis: There is a statistically significant difference in average user ratings for Xbox One and PC.")
else:
    print("We fail to reject the null hypothesis: No significant difference in average user ratings for Xbox One and PC.")

```
